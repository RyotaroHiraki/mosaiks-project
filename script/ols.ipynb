```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

import os
os.chdir('/Users/ryotarohiraki/Desktop/Spring 2026/Capstone/projects')
```

```python
cps14 = pd.read_parquet('dataset/cleaned_dataset/cleaned_cps14_with_nearest_college_imgfeat.parquet')
# cps13 = pd.read_pickle('dataset/cleaned_dataset/cleaned_cps13_with_nearest_college_imgfeat.pkl')
# cps12 = pd.read_pickle('dataset/cleaned_dataset/cleaned_cps12_with_nearest_college_imgfeat.pkl')

cps14.columns
```

```python
#OLS
model = smf.ols('log_wage ~ educ + exp + exp2 + C(PUMA)', data=cps14).fit(cov_type= 'cluster', cov_kwds={'groups': cps14['PUMA']})

print(model.summary())
```

```python
#
img_cols = [c for c in cps14.columns if c.startswith('img_pca_')]

formula = 'log_wage ~ educ + exp + exp2 + female + black +' + " + ".join(img_cols)

model = smf.ols(formula=formula, data=cps14).fit(cov_type= 'cluster', cov_kwds={'groups': cps14['puma']})

print(model.summary())
```
